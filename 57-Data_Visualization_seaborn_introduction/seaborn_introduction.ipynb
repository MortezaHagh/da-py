```python
import matplotlib.pyplot as plt

fig , axes = plt.subplots(1,2, figsize = (9,3.5))
fig.suptitle('website analysis')

axes[0].set_title('views')
axes[0].set_xlabel('day')
axes[0].set_ylabel('count')
axes[0].plot([10,15,12,18,9,13,12],label = 'this week')
axes[0].plot([8,9,13,10,11,14,15],label = 'last week')
axes[0].legend()

axes[1].set_title('new accounts')
axes[1].set_xlabel('day')
axes[1].set_ylabel('count')
axes[1].bar([0,1,2,3,4,5,6],[6,4,4,3,2,4,5])
```

```python
import seaborn as sns
sns.set_theme()

fig , axes = plt.subplots(1,2, figsize = (9,3.5))
fig.suptitle('website analysis')

axes[0].set_title('views')
axes[0].set_xlabel('day')
axes[0].set_ylabel('count')
axes[0].plot([10,15,12,18,9,13,12],label = 'this week')
axes[0].plot([8,9,13,10,11,14,15],label = 'last week')
axes[0].legend()

axes[1].set_title('new accounts')
axes[1].set_xlabel('day')
axes[1].set_ylabel('count')
axes[1].bar([0,1,2,3,4,5,6],[6,4,4,3,2,4,5])
```
